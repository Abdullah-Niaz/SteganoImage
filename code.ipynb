```python
import numpy as np
from PIL import Image
import tensorflow as tf
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input
```

```python
# Load pretrained model
model = MobileNetV2(weights="imagenet", include_top=False, pooling="avg")
```

```python
IMG_SIZE = (224, 224)

def extract_features(img_path):
    img = Image.open(img_path).convert("RGB").resize(IMG_SIZE)
    x = np.array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    features = model.predict(x, verbose=0)
    return features[0]
```

```python
# Reference features
clean_features = extract_features("dataset/clean/image.jpg")
infected_features = extract_features("dataset/infected/image.jpg")
```

```python

# Test image (change this later)
test_image_path = "dataset/infected/image.jpg"
test_features = extract_features(test_image_path)
```

```python
# Cosine similarity
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

sim_clean = cosine_similarity(test_features, clean_features)
sim_infected = cosine_similarity(test_features, infected_features)
```

```python
# Normalize to percentage
total = sim_clean + sim_infected
infected_score = sim_infected / total
clean_score = sim_clean / total
```

```python
label = "Infected" if infected_score > clean_score else "Clean"

print("Prediction:", label)
print(f"Infected: {infected_score*100:.2f}%")
print(f"Clean: {clean_score*100:.2f}%")
```
